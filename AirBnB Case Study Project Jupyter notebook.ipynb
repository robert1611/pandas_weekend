```python
"""
I created two columns, one with the borough population and a second with a per capita population.  
This allows us to see listings on a per capita basis, showing Manhattan is more popular than Manhattan

"""



data = {'Borough': ['Bronx', 'Brookyln', 'Manhattan', 'Queens', ' Staten Island'],
        'Listings': [1091, 20104, 21661, 5666, 373],
        'Population' : [1418000, 2559000, 1628000, 2253000, 476143]
        
       }

df_capita = pd.DataFrame(data)

df_capita['Listings_per_100k_Pop'] = df_capita.Listings / (df_capita.Population / 100000)

df_capita.sort_values(by=['Listings'], ascending = False)

# plt.hist('Listings_per_100k_Pop', density = True, bins = 5)
# plt.ylabel('Listings')
# plt.xlabel('Per 100,000')
# plt.show()

```
